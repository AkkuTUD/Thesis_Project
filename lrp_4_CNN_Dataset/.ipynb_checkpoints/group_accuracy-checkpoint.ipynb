```python
# load two csv files in two pandas dataframes

import pandas as pd
import numpy as np
df_2_classes = pd.read_csv('Animal_Split_2_class.csv',sep=';',index_col=0)
df_9_classes = pd.read_csv('Animal_Split_all_class.csv',sep=';',index_col=0)
```

```python
df_2_classes.info()
```

```python
# extract 4 numbers from the column 'bounding_box'. they are in a list which is an object format. Make 4 new columns with the 4 numbers. They are bounding box coordinates.

df_2_classes['bounding_box'] = df_2_classes['bounding_box'].str.findall(r'\d+')


#df_2_classes['bounding_box'] = df_2_classes['bounding_box'].str.findall(r'\d+')
#df_2_classes['bounding_box'] = df_2_classes['bounding_box'].str.extract('(\d+)').astype(int)
```

```python
df_2_classes['bounding_box'] = df_2_classes['bounding_box'].str.findall(r'\d+')

df_2_classes['x_min'] = pd.to_numeric(df_2_classes['bounding_box'].str[0])
df_2_classes['y_min'] = pd.to_numeric(df_2_classes['bounding_box'].str[1])
df_2_classes['x_max'] = pd.to_numeric(df_2_classes['bounding_box'].str[2])
df_2_classes['y_max'] = pd.to_numeric(df_2_classes['bounding_box'].str[3])
```

```python
df_2_classes['bb_size'] = (df_2_classes['x_max'] - df_2_classes['x_min']) * (df_2_classes['y_max'] - df_2_classes['y_min'])
```

```python
df_2_classes['result'] = True
df_2_classes.loc[(df_2_classes['original_class'] == '1') & (df_2_classes['predicted_class'] == 0),'result'] = False
df_2_classes.loc[(df_2_classes['original_class'] == '0') & (df_2_classes['predicted_class'] == 1),'result'] = False
df_2_classes.loc[(df_2_classes['original_class'] == 'BG'),'result'] = np.nan
```

```python
df_2_classes.sort_values(by='bb_size',ascending=True,inplace=True)
df_2_classes_no_BG = df_2_classes[df_2_classes['original_class'] != 'BG'].copy()
```

```python
df_2_classes_no_BG['size_label'] = 'small'
df_2_classes_no_BG.loc[df_2_classes_no_BG['bb_size'] > 26000,'size_label'] = 'large'
df_2_classes_no_BG.groupby(['size_label','original_class']).agg({'result':'mean','x_min':'count'})
```

```python
pd.crosstab(df_2_classes['original_class'],df_2_classes['predicted_class'])
```

```python
df_2_classes.groupby(['original_class','predicted_class']).agg({'result':'mean','x_min':'count'})
```

```python
# rename classes like in 2-class case
df_9_classes_modified = df_9_classes.copy()
# take only the rows with the original class 1,4,6
df_9_classes_modified = df_9_classes_modified[df_9_classes_modified['original_class'].isin(['1','4','6'])]
# name class 1 as 0
df_9_classes_modified.loc[df_9_classes_modified['original_class'] == '1','original_class'] = '0'
df_9_classes_modified.loc[df_9_classes_modified['predicted_class'] == 1,'predicted_class'] = 0
# name class 4 and 6 as 1
df_9_classes_modified.loc[df_9_classes_modified['original_class'] == '4','original_class'] = '1'
df_9_classes_modified.loc[df_9_classes_modified['original_class'] == '6','original_class'] = '1'
df_9_classes_modified.loc[df_9_classes_modified['predicted_class'] == 4,'predicted_class'] = 1
df_9_classes_modified.loc[df_9_classes_modified['predicted_class'] == 6,'predicted_class'] = 1
```

```python
df_9_classes_modified.head()
```

```python
df_9_classes_modified['bounding_box'] = df_9_classes_modified['bounding_box'].str.findall(r'\d+')

df_9_classes_modified['x_min'] = pd.to_numeric(df_9_classes_modified['bounding_box'].str[0])
df_9_classes_modified['y_min'] = pd.to_numeric(df_9_classes_modified['bounding_box'].str[1])
df_9_classes_modified['x_max'] = pd.to_numeric(df_9_classes_modified['bounding_box'].str[2])
df_9_classes_modified['y_max'] = pd.to_numeric(df_9_classes_modified['bounding_box'].str[3])

df_9_classes_modified['bb_size'] = (df_9_classes_modified['x_max'] - df_9_classes_modified['x_min']) * (df_9_classes_modified['y_max'] - df_9_classes_modified['y_min'])

df_9_classes_modified['result'] = True
df_9_classes_modified.loc[(df_9_classes_modified['original_class'] == '1') & (df_9_classes_modified['predicted_class'] == 0),'result'] = False
df_9_classes_modified.loc[(df_9_classes_modified['original_class'] == '0') & (df_9_classes_modified['predicted_class'] == 1),'result'] = False
#df_9_classes_modified.loc[(df_9_classes_modified['original_class'] == 'BG'),'result'] = np.nan
```

```python
# confusion matrix for df_9_classes_modified
confusion_matrix = pd.crosstab(df_9_classes_modified['original_class'],df_9_classes_modified['predicted_class'])
confusion_matrix
```
